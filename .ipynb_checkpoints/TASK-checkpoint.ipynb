```python
from __future__ import print_function

import os.path

from google.auth.transport.requests import Request
from google.oauth2.credentials import Credentials
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
```

```python
SCOPES = ['https://www.googleapis.com/auth/gmail.readonly', 'https://www.googleapis.com/auth/gmail.compose']
creds = None
# The file token.json stores the user's access and refresh tokens, and is
# created automatically when the authorization flow completes for the first
# time.
if os.path.exists('token.json'):
    
    #creds = Credentials.from_authorized_user_file('token.json')

    creds = Credentials.from_authorized_user_file('token.json', SCOPES)
# If there are no (valid) credentials available, let the user log in.
if not creds or not creds.valid:
    if creds and creds.expired and creds.refresh_token:
        creds.refresh(Request())
    else:
        flow = InstalledAppFlow.from_client_secrets_file(
            'credentials_web2.json', SCOPES)
        creds = flow.run_local_server(port=0)
    # Save the credentials for the next run
    with open('token.json', 'w') as token:
        token.write(creds.to_json())
```

```python


# If modifying these scopes, delete the file token.json.


def main():
    """Shows basic usage of the Gmail API.
    Lists the user's Gmail labels.
    """


    try:
        # Call the Gmail API
        service = build('gmail', 'v1', credentials=creds)
        results = service.users().labels().list(userId='me').execute()
        labels = results.get('labels', [])

        if not labels:
            print('No labels found.')
            return
        print('Labels:')
        for label in labels:
            print(label['name'])

    except HttpError as error:
        # TODO(developer) - Handle errors from gmail API.
        print(f'An error occurred: {error}')


if __name__ == '__main__':
    main()
```

```python
import os
import base64
from google.oauth2.credentials import Credentials
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError




def authenticate_google_api():
    """
    Authenticates the Google API using the client ID and secret in the JSON file.
    """
    # Set the path to the client ID JSON file
    CLIENT_SECRET_FILE = 'credentials_desktop3.json'

    # Define the credentials for the OAuth 2.0 flow
    flow = InstalledAppFlow.from_client_secrets_file(
        CLIENT_SECRET_FILE, SCOPES)
    credentials = flow.run_local_server(port=0)

    return credentials

# Authenticate the Google API
credentials = authenticate_google_api()

        
def send_email(to, subject, body):
    """
    Sends an email with the given subject and body to the specified recipient.
    """
    try:

        # Build the Gmail API client
        service = build('gmail', 'v1', credentials=credentials)

        # Define the email message
        message = create_message(to, subject, body)

        # Send the message using the Gmail API
        send_message = (service.users().messages().send(userId="me", body=message)
                        .execute())
        print(F'Sent message to {to} Message Id: {send_message["id"]}')

    except HttpError as error:
        print(F'An error occurred: {error}')
        send_message = None

    return send_message


def create_message(to, subject, body):
    """
    Creates a message for an email.
    """
    message = (f'From: Your Name <mehtab.karmani@gmail.com>\r\n'
               f'To: {to}\r\n'
               f'Subject: {subject}\r\n\r\n'
               f'{body}')
    message_bytes = message.encode('utf-8')
    encoded_message = base64.urlsafe_b64encode(message_bytes).decode('utf-8')
    return {'raw': encoded_message}



```

```python
# Example usage
send_email('mehtab.karmani@gmail.com', 'Test email', 'This is a test email.')
```

```python
import base64
from google.oauth2.credentials import Credentials
from googleapiclient.discovery import build


data = []


def search_emails_with_keywords(keywords):
    """
    Searches for emails with the specified keywords in the subject or body.
    """
    try:
        # Authenticate the Google API
        #credentials = authenticate_google_api()

        # Build the Gmail API client
        service = build('gmail', 'v1', credentials=credentials)

        # Define the search query
        query = ''
        for keyword in keywords:
            query += f'subject:{keyword} OR '
            query += f'body:{keyword} OR '
        query = query[:-4]  # Remove the last ' OR '

        # Search for emails matching the query
        messages = service.users().messages().list(
            userId='me', q=query).execute().get('messages', [])

        # Print the subject and snippet of each matching email
        for message in messages:
            msg = service.users().messages().get(userId='me', id=message['id']).execute()
            
            subject = [header['value'] for header in msg['payload']['headers']
            #           if header['name'] == 'Subject'][0]
            snippet = msg['snippet']
            data.append({'subject':subject, 'msg':snippet})
            #print(f'Subject: {subject}\Body: {snippet}\n\n')

    except HttpError as error:
        print(F'An error occurred: {error}')
        messages = []

    return messages


# Example usage
keywords = ['DATA ENGINEER']
search_emails_with_keywords(keywords)
```
